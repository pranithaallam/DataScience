# **Standardization of Numerical Variables**

```python
numerical_columns = df.select_dtypes(include=['int64']).columns
print("Numerical Columns:", numerical_columns)
```

**Concept of standardization:**Also kmown a z-score.It is a scaling method where the values are centered around mean with a unit standard deviation.It means if we will calculate mean and standard deviation of standard scores ot will be 0 and 1.
Range is -3 to 3.


```python
means = df[numerical_columns].mean()
stds = df[numerical_columns].std()
df_standardized = df.copy()
df_standardized[numerical_columns] = (df[numerical_columns] - means) / stds
df_standardized[numerical_columns]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 6))

# Plot Standardized Age
sns.histplot(df['Volume'], kde=True, ax=axes[0])
axes[0].set_title('Standardized Volume Distribution')
axes[0].set_xlabel('Standardized Volume')
axes[0].set_ylabel('Frequency')

# Plot Standardized Salary
sns.histplot(df['Avg Price'], kde=True, ax=axes[1])
axes[1].set_title('Standardized Avg Price Distribution')
axes[1].set_xlabel('Standardized Avg Price')
axes[1].set_ylabel('Frequency')

sns.histplot(df['Total Sales Value'], kde=True, ax=axes[2])
axes[2].set_title('Standardized Total Sales Value Distribution')
axes[2].set_xlabel('Standardized Total Sales Value')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
#compare the graphs of
```
