# **Dimensionality Reduction with PCA**

# Standardizing the features

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data#representing scaled data
```

# Implementing PCA

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Apply PCA
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

# Optimal number of principal components

```python
# Scree plot for explained variance
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.show()

#cumulative explained variance.
explained_variance = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(explained_variance >= 0.95) + 1
print(f'Number of components that explain 95% variance: {num_components}')
```

# Transforming original data

```python
pca = PCA(n_components=num_components)
pca_transformed = pca.fit_transform(scaled_data)
```

# **Clustering with Original Data**

# K-means to the original dataset

```python
from sklearn.cluster import KMeans

# Apply K-Means on the original dataset
kmeans = KMeans(n_clusters=3, random_state=42)
original_labels = kmeans.fit_predict(scaled_data)
```

# Visualize clustering results

```python
reduced_data = kmeans.fit_transform(scaled_data)

plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=original_labels, cmap='viridis')
plt.title('K-Means Clustering on Original Data')
```

# Hieracheal clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Hierarchical clustering
linked = linkage(scaled_data, method='ward')

# Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram for Hierarchical Clustering')
plt.show()
```

```python
# Apply Agglomerative Clustering,taking clusters as 2
hc = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
hc_labels = hc.fit_predict(scaled_data)
```

# Clustering performance using metrics such as silhouette score.

```python
from sklearn.metrics import silhouette_score

silhouette_originalkm = silhouette_score(scaled_data, original_labels)
print(f'Silhouette Score (Original Data): {silhouette_originalkm}')


silhouette_originalhc = silhouette_score(scaled_data, hc_labels)
print(f'Silhouette Score (Original Data): {silhouette_originalhc}')
```

# Practical implications of using PCA and clustering in data analysis

**PCA**:

Reducing dimensions while preserving maximum variance.

New features (principal components) are abstract.

Requires standardization of data before application of PCA.

Exploratory, data preparation for other models.

**Clustering**:

Grouping similar data points into clusters.

Group assignments are easy to interpret.

Sensitive to scaling but more flexible depending on algorithm.

Exploratory and predictive, useful for segmentation
