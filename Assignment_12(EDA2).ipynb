# One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
categorical_one_hot = df.select_dtypes(include=['object']).nunique() < 5
data = pd.get_dummies(df, columns=categorical_one_hot[categorical_one_hot].index.tolist(), drop_first=True)
```

```python
data
```

# Label Encoding

```python
label_encoder={}
categorical_label1=data.select_dtypes(include=['object']).nunique()>=5
for column in categorical_label1[categorical_label1].index:
  le=LabelEncoder()
  data[column]=le.fit_transform(data[column])
  label_encoder[column]=le
```

# **Pros and cons of One-Hot Encoding and Label Encoding**

# One-Hot Encoding:
**Pros**:

One-hot encoding does not assume any order between categories.It's suitable for unordered categorical variables.

It works well for algorithms like decision tree,random forest,neural networks that don't assume any relationship between feature values.

All categories are represented without imposing any ranking or numerical assumptions,preserving the full information about the categorical variable.

This technique avoids introducing bias into models that might misinterpret numeric encoding as ordinal.

**Cons**:

For features with a large number of categories,it can create many columns,increasing the dimensionality of dataset.

In case of where many categorical variables are present,the resulting dataset can become unnecessarily complex.


# Label Encoding:

**Pros**:

Label Encoding is simple to implement and requires only one column,maintaining the dataset original size.

It doesn't increase dimensionality and is memory-efficient,making it suitable for algorithm that handle large dataset.

It appropriate as it preserves the ordinal relationships between categories.

**Cons**:

Ordinal issues in nominal data like some models may interpret higher numerical values as having more importance,which can introduce bias when there is no natural ordering in data.

For features with a large number of categories it can make difficult for models to distinguish between categories.
