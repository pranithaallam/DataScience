# **Recommendation System**

```python
# Computing cosine similarity between all anime
cosine_sim_matrix = cosine_similarity(anime_df)

# Function to recommend similar anime based on cosine similarity
def recommend_anime(anime_title, num_recommendations=5, similarity_threshold=0.5):
    # Ensure the anime_title exists in the index
    if anime_title not in anime_df.index:
        return f"Anime titled '{anime_title}' not found in the dataset."

    # Find the index of the anime title
    anime_idx = anime_df.index.get_loc(anime_title)

    # Get similarity scores for all anime compared to the target anime
    similarity_scores = list(enumerate(cosine_sim_matrix[anime_idx]))

    # Sort by similarity scores in descending order
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Filter based on the similarity threshold
    filtered_scores = [(i, score) for i, score in similarity_scores[1:] if score >= similarity_threshold]

    # Get the indices of the most similar anime, up to the specified number of recommendations
    similar_anime_indices = [i for i, score in filtered_scores[:num_recommendations]]

    # If no anime meet the similarity threshold, return a message
    if not similar_anime_indices:
        return f"No anime found with similarity greater than {similarity_threshold}."

    # Return the titles of the recommended anime
    return anime_df.index[similar_anime_indices].values
```

# **Evaluation**

```python
# Spliting data into training and test sets (for evaluation)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Function to evaluate precision, recall, F1-score (using cosine similarity)
def evaluate_recommendations(test_data, num_recommendations=5):
    y_true = []  # List of true anime (1 if true anime is in recommendations, 0 otherwise)
    y_pred = []  # List of predicted labels (same shape as y_true)

    for index, row in test_data.iterrows():
        true_anime = row['name']
        recommended_anime = recommend_anime(true_anime, num_recommendations)

        # If the true anime is in the recommended list, mark it as 1, else 0
        if true_anime in recommended_anime:
            y_true.append(1)
            y_pred.append(1)
        else:
            y_true.append(1)
            y_pred.append(0)
#Analyzing the performance
    # Calculate precision, recall, and F1-score
    precision = precision_score(y_true, y_pred, average='binary')
    recall = recall_score(y_true, y_pred, average='binary')
    f1 = f1_score(y_true, y_pred, average='binary')

    return precision, recall, f1

# Evaluating the system
precision, recall, f1 = evaluate_recommendations(test_data)
print(f'Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
```

# **Difference between user-based and item-based collaborative filtering:**


**User-based filtering:** Recommends items by identifying users with similar behavior (i.e., similar item preferences) to the target user and recommending items liked by similar users.


**Item-based filtering:** Recommends items based on their similarity to items the target user has already interacted with. Instead of finding similar users, it finds similar items.

# **Collaborative filtering, and how does it works:**
Collaborative filteering is one of the type of recommendation system.

Collaborative Filtering recommends items based on similarity measures between users and/or items. The basic assumption behind the algorithm is that users with similar interests have common preferences.

There are two main types of collaborative filtering:

1.User-based Collaborative Filtering

2.Item-based Collaborative Filtering

1. **User-based Collaborative Filtering:**
In user-based collaborative filtering, the idea is to find users who are similar to the target user based on their past behavior (e.g., ratings of movies or anime). Then, based on the preferences of these similar users, the system recommends items that the target user hasn't interacted with but that similar users have liked.

  Working:
For a given target user, calculate the similarity between them and other users based on their past ratings. The similarity can be measured using metrics like cosine similarity or Pearson correlation.

  Once similar users are found, the system looks at the items (e.g., movies, anime) that those similar users have rated highly but that the target user has not yet seen. These items are then recommended to the target user.

  2**.Item-based Collaborative Filtering:**

  In item-based collaborative filtering, the focus is on finding similarities between items rather than users. The idea is that if a user has liked certain items in the past, they will like similar items in the future.

  Working:

 For each item (e.g., anime), calculate how similar it is to other items based on user ratings. Again, metrics like cosine similarity or Pearson correlation can be used to measure similarity.

 Based on the user's past preferences, recommend items that are most similar to the ones the user has rated highly.
