```python
#importing required libraries
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#loading dataset
data = pd.read_csv('/content/OnlineRetail.csv')
```

```python
data.head()#first 5 rows of dataset
```

# **Data Preprocessing**

```python
data.info()#checking datatypes, null values,number of columns
```

```python
data.describe()#summary statistics of data
```

# Handling missing values

```python
#checking null values,not removing the null because dataset will remain with no dataset if i remove null and also replacement causes errors,so keeping it same.
data.isnull().sum()
```

# Removing duplicates

```python
data.drop_duplicates(inplace=True)
```

# Converting the data to appropriate format

```python
df=pd.get_dummies(data)
df.head()
```

```python
df.describe()#summary statistics of data
```

# **Association Rule Mining**

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)#df-dataframe, min_support default=0.5
# min_support is a float between 0 and 1
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1) # min_threshold is Confidence
rules
rules.sort_values('lift',ascending = True)[0:20]
```

```python
rules[rules.lift>1]#results
```

# Some limitations or challenges of Association rules mining

The dataset may contain a large number of items, leading to a combinatorial explosion of possible itemsets.

Setting an appropriate support threshold is challenging; too high may miss significant rules, and too low may result in noise.

The rules generated may not always be intuitive or actionable, making interpretation subjective.

Customer purchasing behavior changes over time, and static models may not capture these dynamics effectively.
